# Searching Persons

```python
import spacy
nlp = spacy.load('en')
doc = nlp(open('input.txt').read())
```

```python
for person in doc.ents:
    if(person.label_ == 'PERSON'):
        print(person.text)
```

# Serching Dates

```python
for date in doc.ents:
    if(date.label_ == 'DATE'):
        print(date.text)
```
